```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
omni = pd.read_csv('OMNI_XYZ.csv')
omni['date'] = pd.to_datetime(omni['date'], format='%Y-%m-%d %H:%M:%S')
omni
```

```python
themis = pd.read_csv('themis.csv')
themis
```

```python
# transform the themis['TIMESTAMP']'s format to omni['TIMESTAMP']'s format like YYYY-MM-DD HH:MM:SS
themis['TIMESTAMP'] = pd.to_datetime(themis['TIMESTAMP'], format='%Y-%m-%d %H:%M:%S')
themis['TIMESTAMP'] = themis['TIMESTAMP'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
themis = themis.drop_duplicates(subset=['TIMESTAMP'], keep='first')
themis = themis.rename(columns={'TIMESTAMP': 'date'})
themis['date'] = pd.to_datetime(themis['date'], format='%Y-%m-%d %H:%M:%S')
themis
```

```python
omni.dtypes, themis.dtypes
```

```python
omni_indexed = omni.set_index('date')
omni_indexed
```

```python
results = []

for index, row in themis.iterrows():
    timestamp = row['date']
    start_time = timestamp - pd.Timedelta(minutes=5)
    end_time = timestamp + pd.Timedelta(minutes=5)
    omni_subset = omni_indexed.loc[start_time:end_time]
    if len(omni_subset) > 0:
        bx_avg = omni_subset['bx'].mean()
        by_avg = omni_subset['by'].mean()
        bz_avg = omni_subset['bz'].mean()
        dp_avg = omni_subset['Dp'].mean()
        angle =  omni_subset['angle'].mean()
        vt = omni_subset['sw_vt'].mean()
        result = {'date': timestamp, 'Bx': bx_avg, 'By': by_avg, 'Bz': bz_avg, 'Dp': dp_avg, 'angle' : angle, 'vt' : vt}
        results.append(result)
```

```python
results = pd.DataFrame(results)
results
```

```python
# combine themis and results
themis = themis.set_index('date')
themis = themis.join(results.set_index('date'))
themis
```

```python
themis.to_csv('themis_omni.csv', index=True)
```

# Geo-Tail

```python
# since geo-tails UT is in a format of 2.2267, 2.2400, 13.0467 and we like to transform it to HH:MM:SS
geo_tail = pd.read_csv('deep_tail.csv')
def transform_UT_to_HHMMSS(UT):
    hour = int(UT)
    minute = int((UT - hour) * 60)
    second = int((((UT - hour) * 60) - minute) * 60)
    return str(hour).zfill(2) + ':' + str(minute).zfill(2) + ':' + str(second).zfill(2)

geo_tail['UT'] = geo_tail['UT'].apply(transform_UT_to_HHMMSS)
geo_tail['date'] = geo_tail['YYYYMMDD'].astype(str) + ' ' + geo_tail['UT'].astype(str)
geo_tail['date'] = pd.to_datetime(geo_tail['date'], format='%Y%m%d %H:%M:%S')
geo_tail = geo_tail.drop(columns=['YYYYMMDD', 'UT'])
geo_tail
```

```python
# use the same approach to combine omni and geo_tail
omni_indexed = omni.set_index('date')
omni_indexed
```

```python
results = []

for index, row in geo_tail.iterrows():
    timestamp = row['date']
    start_time = timestamp - pd.Timedelta(minutes=5)
    end_time = timestamp + pd.Timedelta(minutes=5)
    omni_subset = omni_indexed.loc[start_time:end_time]
    if len(omni_subset) > 0:
        bx_avg = omni_subset['bx'].mean()
        by_avg = omni_subset['by'].mean()
        bz_avg = omni_subset['bz'].mean()
        dp_avg = omni_subset['Dp'].mean()
        angle =  omni_subset['angle'].mean()
        vt = omni_subset['sw_vt'].mean()
        result = {'date': timestamp, 'bx_avg': bx_avg, 'by_avg': by_avg, 'bz_avg': bz_avg, 'Dp': dp_avg, 'angle' : angle, 'vt' : vt}
        results.append(result)
```

```python
results = pd.DataFrame(results)
results
```

```python
geo_tail = geo_tail.set_index('date')
geo_tail = geo_tail.join(results.set_index('date'))
geo_tail
```

```python
geo_tail.drop(columns=['Vx', 'Vy', 'Vz', 'Bx', 'By', 'Bz', 'N', 'T'], inplace=True)
geo_tail
```

```python
geo_tail.rename(columns={'bx_avg': 'Bx', 'by_avg' : 'By', 'bz_avg' : 'Bz', 'dp_avg' : 'Dp'}, inplace=True)
geo_tail
```

```python
geo_tail.to_csv('geotail_omni.csv', index=True)
```

```python
# concatenate the geo_tail and themis
themis = pd.read_csv('themis_omni.csv')
themis.drop(['PROBE'], axis=1, inplace=True)
themis
```

```python
geo_tail = pd.read_csv('geotail_omni.csv')
geo_tail.rename(columns={'XGSE':'X_GSE', 'YGSE' : 'Y_GSE', 'ZGSE' : 'Z_GSE', 'XGSM' : 'X_GSM'\
    , 'YGSM' : 'Y_GSM', 'ZGSM' : 'Z_GSM'}, inplace=True)
geo_tail
```

```python
# concatenate themis and geo_tail
themis_geo_tail = pd.concat([themis, geo_tail])
themis_geo_tail
```

```python
themis_geo_tail
```

```python
themis_geo_tail['phi'] = np.arctan2(themis_geo_tail['Y_GSM'], themis_geo_tail['Z_GSM'])
themis_geo_tail
```

```python
themis_geo_tail['X_GSM_r'] = themis_geo_tail['X_GSE']
themis_geo_tail['Y_GSM_r'] = -(themis_geo_tail['Y_GSE']*(np.cos(themis_geo_tail['angle'])) + \
    themis_geo_tail['Z_GSE']*np.sin((themis_geo_tail['angle'])))
themis_geo_tail['Z_GSM_r'] = -themis_geo_tail['Y_GSE']*np.sin(themis_geo_tail['angle'])*np.sin(themis_geo_tail['phi']) + \
    themis_geo_tail['Z_GSE']*np.cos(themis_geo_tail['angle']) - themis_geo_tail['Y_GSE']*np.sin(themis_geo_tail['angle'])*np.cos(themis_geo_tail['phi'])
    
themis_geo_tail
```

```python
# reset index
themis_geo_tail.reset_index(inplace=True)
themis_geo_tail.drop(columns=['index'], inplace=True)
themis_geo_tail
```

```python
# save to csv
themis_geo_tail.to_csv('themis_geotail_omni_raw.csv', index=True)
```

```python
data['R'] = (data['Y_GSM_r']**2 + data['Z_GSM_r']**2)**0.5  # R = sqrt(Y^2 + Z^2)
data['r'] = (data['X_GSM_r']**2 + data['Y_GSM_r']**2 + data['Z_GSM_r']**2)**0.5  # r = sqrt(X^2 + Y^2 + Z^2)
data['theta'] = np.arctan2(data['Y_GSM_r'], data['X_GSM_r'])
data
```
